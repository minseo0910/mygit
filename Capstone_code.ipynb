* heartrate.csv 전처리

```python
import os
import pandas as pd

# 원본 데이터 경로와 저장할 경로 설정
input_dir = r'C:\Users\user\Desktop\Woo LAB\03Fitbit\Fitbit_data'
output_dir = r'C:\Users\user\Desktop\Woo LAB\03Fitbit\Fitbit_clean'

# output_dir 경로에 각 사용자 폴더 생성
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 사용자 폴더 리스트 가져오기
user_folders = [f for f in os.listdir(input_dir) if os.path.isdir(os.path.join(input_dir, f))]

for user_id in user_folders:
    user_folder = os.path.join(input_dir, user_id)
    heartrate_file = os.path.join(user_folder, 'heartrate.csv')

    # heartrate.csv 파일이 존재하는지 확인
    if os.path.exists(heartrate_file):
        # 데이터 로드
        df = pd.read_csv(heartrate_file)

        # dateTime에서 날짜만 추출
        df['date'] = pd.to_datetime(df['dateTime']).dt.date

        # 날짜별로 value.bpm의 평균과 총합 계산
        bpm_summary = df.groupby('date').agg(
            avg_bpm=('value.bpm', 'mean'),
            sum_bpm=('value.bpm', 'sum')
        ).reset_index()

        # 저장할 폴더 경로 생성
        user_output_folder = os.path.join(output_dir, user_id)
        if not os.path.exists(user_output_folder):
            os.makedirs(user_output_folder)

        # 새로운 파일 경로
        output_file = os.path.join(user_output_folder, 'heartrate.csv')

        # 새로운 데이터 저장
        bpm_summary.to_csv(output_file, index=False)

        print(f"Processed and saved for user {user_id}")

print("All files processed successfully.")
```

exercise 전처리

```python
import os
import pandas as pd

# 원본 데이터 경로와 저장할 경로 설정
input_dir = r'C:\Users\user\Desktop\Woo LAB\03Fitbit\Fitbit_data'
output_dir = r'C:\Users\user\Desktop\Woo LAB\03Fitbit\Fitbit_clean'

# 선택할 열 목록
selected_columns = [
    'activityTypeId', 'averageHeartRate', 'calories', 'activeDuration', 'steps',
    'sedentary', 'lightly', 'fairly', 'very', 'out_of_range', 'fat_burn', 'cardio', 'peak', 'startTime'
]

# output_dir 경로에 각 사용자 폴더 생성
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 사용자 폴더 리스트 가져오기
user_folders = [f for f in os.listdir(input_dir) if os.path.isdir(os.path.join(input_dir, f))]

for user_id in user_folders:
    user_folder = os.path.join(input_dir, user_id)
    exercise_file = os.path.join(user_folder, 'exercise.csv')

    # exercise.csv 파일이 존재하는지 확인
    if os.path.exists(exercise_file):
        # 데이터 로드
        df = pd.read_csv(exercise_file)

        # startTime에서 날짜만 추출
        df['startTime'] = pd.to_datetime(df['startTime']).dt.date

        # 'sedentary', 'lightly', 'fairly', 'very'에서 'minutes' 값만 추출
        for col in ['sedentary', 'lightly', 'fairly', 'very']:
            df[col] = df[col].apply(lambda x: eval(x)['minutes'] if pd.notnull(x) else None)

        # 'out_of_range', 'fat_burn', 'cardio', 'peak'에서 'caloriesOut' 값만 추출하고 소수점 제거
        for col in ['out_of_range', 'fat_burn', 'cardio', 'peak']:
            df[col] = df[col].apply(lambda x: int(eval(x).get('caloriesOut', 0)) if pd.notnull(x) else None)

        # 필요한 열만 선택
        df_selected = df[selected_columns]

        # 저장할 폴더 경로 생성
        user_output_folder = os.path.join(output_dir, user_id)
        if not os.path.exists(user_output_folder):
            os.makedirs(user_output_folder)

        # 새로운 파일 경로
        output_file = os.path.join(user_output_folder, 'exercise.csv')

        # 새로운 데이터 저장
        df_selected.to_csv(output_file, index=False)

        print(f"Processed and saved exercise data for user {user_id}")

print("All exercise files processed successfully.")
```

**각 사용자마다 날짜별로 파일 통합**

```python
import os
import pandas as pd

# 데이터 경로 설정
input_dir = r'C:\Users\user\Desktop\Woo LAB\03Fitbit\Fitbit_clean'

# 사용자 폴더 리스트 가져오기
user_folders = [f for f in os.listdir(input_dir) if os.path.isdir(os.path.join(input_dir, f))]

# 파일 병합 함수 정의
def merge_user_files(user_folder):
    # 파일 경로 설정
    files = {
        'activities': os.path.join(user_folder, 'activities.csv'),
        'calories': os.path.join(user_folder, 'calories.csv'),
        'heartrate': os.path.join(user_folder, 'heartrate.csv'),
        'steps': os.path.join(user_folder, 'steps.csv'),
        'sleep': os.path.join(user_folder, 'sleep.csv'),
        'exercise': os.path.join(user_folder, 'exercise.csv')
    }
    
    # 각 파일을 읽어들임, 없는 파일은 무시
    dfs = {}
    for key, file_path in files.items():
        if os.path.exists(file_path):
            df = pd.read_csv(file_path)
            if key == 'sleep':
                df.rename(columns={'dateTime': 'date'}, inplace=True)  # 'sleep' 파일의 'dateTime'을 'date'로 변경
            elif key == 'exercise':
                df.rename(columns={'startTime': 'date'}, inplace=True)  # 'exercise' 파일의 'startTime'을 'date'로 변경
            dfs[key] = df

    # 날짜를 기준으로 병합
    combined_df = None
    for df in dfs.values():
        if combined_df is None:
            combined_df = df
        else:
            combined_df = pd.merge(combined_df, df, on='date', how='outer')

    # 날짜 오름차순 정렬
    combined_df = combined_df.sort_values(by='date')

    return combined_df

# 각 사용자에 대해 병합된 파일 저장
for user_id in user_folders:
    user_folder = os.path.join(input_dir, user_id)

    # 병합된 데이터 가져오기
    combined_data = merge_user_files(user_folder)

    # 새로운 파일 경로 설정
    output_file = os.path.join(user_folder, 'Combine.csv')

    # 병합된 데이터 저장
    combined_data.to_csv(output_file, index=False)

    print(f"Combined data saved for user {user_id}")

print("All users' data combined successfully.")
```

```python
import os
import pandas as pd

# 데이터 경로 설정
input_dir = r'C:\Users\user\Desktop\Woo LAB\03Fitbit\Fitbit_clean'

# 사용자 폴더 리스트 가져오기
user_folders = [f for f in os.listdir(input_dir) if os.path.isdir(os.path.join(input_dir, f))]

# 파일 전처리 함수 정의
def add_time_steps(user_folder):
    file_path = os.path.join(user_folder, 'Combine.csv')

    if os.path.exists(file_path):
        # 데이터 로드
        df = pd.read_csv(file_path)

        # date 열 제외한 44개 변수 선택
        columns = [col for col in df.columns if col != 'date']

        # 각 시점(t-1 ~ t-7)을 위한 열 추가
        for t in range(1, 8):  # t-1 to t-7
            shifted_df = df[columns].shift(t)
            shifted_df.columns = [f'P{t}_{col}' for col in columns]
            df = pd.concat([df, shifted_df], axis=1)

        # 새로운 파일 경로 설정 (Combine_T.csv로 저장)
        output_file = os.path.join(user_folder, 'Combine_T.csv')

        # 처리된 데이터 저장
        df.to_csv(output_file, index=False)

        print(f"Added time-step variables and saved Combine_T.csv for user {os.path.basename(user_folder)}")

# 각 사용자에 대해 Combine 파일에 시점 변수 추가 후 저장
for user_id in user_folders:
    user_folder = os.path.join(input_dir, user_id)
    add_time_steps(user_folder)

print("All Combine files processed successfully with time-step variables added and saved as Combine_T.csv.")
```

```python
## 과거 데이터만 가지고 보간법 후 KNN으로 전처리
import pandas as pd
from sklearn.impute import KNNImputer

# 데이터 로드
file_path = 'C:/Users/user/Desktop/Woo LAB/03Fitbit/ALL_user.csv'
data = pd.read_csv(file_path)

# 삭제할 사용자 리스트
users_to_remove = ['S004', 'S026', 'S082', 'S090', 'S503', 'S015', 'S049']
data_filtered = data[~data['userID'].isin(users_to_remove)]

# 사용자별로 그룹화
grouped = data_filtered.groupby('userID')

# 빈 데이터프레임 생성
filled_data = pd.DataFrame()

# 각 사용자별로 결측치를 처리
for user, group in grouped:
    # userID와 date 컬럼을 제외한 나머지 숫자 컬럼만 추출
    user_data_numeric = group.drop(columns=['userID', 'date'])
    
    # 과거 데이터로 보간법을 적용하여 결측치를 채움 (forward fill)
    user_data_filled_numeric = user_data_numeric.fillna(method='ffill')
    
    # KNN Imputer를 사용하여 남은 결측치를 채움
    knn_imputer = KNNImputer(n_neighbors=5)
    user_data_knn_filled = pd.DataFrame(knn_imputer.fit_transform(user_data_filled_numeric), columns=user_data_numeric.columns)
    
    # userID와 date 컬럼을 다시 추가
    user_data_filled = pd.concat([group[['userID', 'date']].reset_index(drop=True), user_data_knn_filled], axis=1)
    
    # 채워진 데이터를 빈 데이터프레임에 추가
    filled_data = pd.concat([filled_data, user_data_filled], ignore_index=True)

# 결측치가 채워진 전체 데이터 저장
knn_filled_path = 'C:/Users/user/Desktop/Woo LAB/03Fitbit/ALL_user_eda.csv'
filled_data.to_csv(knn_filled_path, index=False)

print("과거 데이터를 사용한 KNN 기반 사용자별 결측치 채우기 완료.")
```

```python
## t 시점
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import tensorflow as tf

# MAPE 계산 함수 정의
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# WaveBound 손실 함수 정의
def wavebound_loss(y_true, y_pred, target_pred, epsilon=0.1):
    loss = tf.abs(mean_squared_error(y_true, y_pred) - (mean_squared_error(y_true, target_pred) - epsilon))
    return loss + (mean_squared_error(y_true, target_pred) - epsilon)

# 타겟 네트워크 업데이트 (EMA 사용)
def update_target_weights(source_model, target_model, alpha=0.99):
    source_weights = np.array(source_model.get_weights())
    target_weights = np.array(target_model.get_weights())
    new_weights = alpha * target_weights + (1 - alpha) * source_weights
    target_model.set_weights(new_weights)

# 파일 경로 및 데이터 로드
file_path = r'C:\Users\user\Desktop\Woo LAB\03Fitbit\ALL_user_eda.csv'
df = pd.read_csv(file_path)

# t 시점 변수 (20개) 선택
selected_columns_t = [
    'activities', 'calories_avg', 'calories_sum', 'HR_avg', 'HR_sum', 'steps_avg', 'steps_sum',
    'activityType', 'exercise_HR', 'exercise_calories', 'activeDuration', 'exercise_steps',
    'sedentary', 'lightly', 'fairly', 'very', 'out_of_range', 'fat_burn', 'cardio', 'peak'
]

# 입력 변수(X)와 타겟 변수(y) 설정
X = df[selected_columns_t].values
y = df['overall_score'].values

# 데이터 정규화
scaler = MinMaxScaler(feature_range=(0, 1))
X_scaled = scaler.fit_transform(X)

# 데이터 7:3으로 나누기
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, shuffle=False)

# LSTM 입력을 위해 데이터의 형태 변경 (samples, timesteps, features)
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))

# LSTM 모델 생성 함수
def build_lstm_model(input_shape):
    model = Sequential()
    model.add(LSTM(64, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(LSTM(64))
    model.add(Dropout(0.2))
    model.add(Dense(1))
    return model

input_shape = (1, X_train.shape[2])

# 소스 및 타겟 모델 생성
source_model = build_lstm_model(input_shape)
target_model = build_lstm_model(input_shape)

# 소스 모델 컴파일 (기본 MSE 사용)
optimizer = Adam(learning_rate=0.001)
source_model.compile(optimizer=optimizer, loss='mean_squared_error')

# WaveBound를 적용한 학습 루프
epochs = 100
batch_size = 32
epsilon = 0.1  # WaveBound의 하이퍼파라미터

for epoch in range(epochs):
    # 소스 모델 학습
    source_model.fit(X_train, y_train, epochs=1, batch_size=batch_size, validation_data=(X_test, y_test), verbose=1)
    
    # 타겟 모델로부터 예측값 생성 (WaveBound 손실 함수에 사용)
    target_pred = target_model.predict(X_train)
    
    # 타겟 모델 업데이트 (EMA 사용)
    update_target_weights(source_model, target_model)
    
    # WaveBound 손실 계산 및 소스 모델 업데이트
    loss = wavebound_loss(y_train, source_model.predict(X_train), target_pred, epsilon=epsilon)
    
    print(f"Epoch {epoch+1}/{epochs}, Loss: {loss}")

# 예측 수행
y_pred = source_model.predict(X_test)

# 성능 평가
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
r_squared = r2_score(y_test, y_pred)

print(f"WaveBound LSTM Model Performance (t 시점):")
print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
print(f"MAPE: {mape}%") 
print(f"R-squared: {r_squared}")
```

```python
## t-7 시점
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import tensorflow as tf

# MAPE 계산 함수 정의
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# WaveBound 손실 함수 정의
def wavebound_loss(y_true, y_pred, target_pred, epsilon=0.1):
    loss = tf.abs(mean_squared_error(y_true, y_pred) - (mean_squared_error(y_true, target_pred) - epsilon))
    return loss + (mean_squared_error(y_true, target_pred) - epsilon)

# 타겟 네트워크 업데이트 (EMA 사용)
def update_target_weights(source_model, target_model, alpha=0.99):
    source_weights = np.array(source_model.get_weights())
    target_weights = np.array(target_model.get_weights())
    new_weights = alpha * target_weights + (1 - alpha) * source_weights
    target_model.set_weights(new_weights)

# 파일 경로 및 데이터 로드
file_path = r'C:\Users\user\Desktop\Woo LAB\03Fitbit\ALL_user_eda.csv'
df = pd.read_csv(file_path)

# t 시점 + t-1 시점 + t-2 시점 + t-3 시점 + t-4 시점 + t-5 시점 + t-6 시점 + t-7 시점 변수 선택
selected_columns_t = [
    'activities', 'calories_avg', 'calories_sum', 'HR_avg', 'HR_sum', 'steps_avg', 'steps_sum',
    'activityType', 'exercise_HR', 'exercise_calories', 'activeDuration', 'exercise_steps',
    'sedentary', 'lightly', 'fairly', 'very', 'out_of_range', 'fat_burn', 'cardio', 'peak'
]

selected_columns_t1 = [col for col in df.columns if col.startswith('P1_')]
selected_columns_t2 = [col for col in df.columns if col.startswith('P2_')]
selected_columns_t3 = [col for col in df.columns if col.startswith('P3_')]
selected_columns_t4 = [col for col in df.columns if col.startswith('P4_')]
selected_columns_t5 = [col for col in df.columns if col.startswith('P5_')]
selected_columns_t6 = [col for col in df.columns if col.startswith('P6_')]
selected_columns_t7 = [col for col in df.columns if col.startswith('P7_')]

# 전체 328개 변수 설정
selected_columns = selected_columns_t + selected_columns_t1 + selected_columns_t2 + selected_columns_t3 + selected_columns_t4 + selected_columns_t5 + selected_columns_t6 + selected_columns_t7

# 입력 변수(X)와 타겟 변수(y) 설정
X = df[selected_columns].values
y = df['overall_score'].values

# 데이터 정규화
scaler = MinMaxScaler(feature_range=(0, 1))
X_scaled = scaler.fit_transform(X)

# 데이터 7:3으로 나누기
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, shuffle=False)

# LSTM 입력을 위해 데이터의 형태 변경 (samples, timesteps, features)
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))

# LSTM 모델 생성 함수
def build_lstm_model(input_shape):
    model = Sequential()
    model.add(LSTM(64, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(LSTM(64))
    model.add(Dropout(0.2))
    model.add(Dense(1))
    return model

input_shape = (1, X_train.shape[2])

# 소스 및 타겟 모델 생성
source_model = build_lstm_model(input_shape)
target_model = build_lstm_model(input_shape)

# 소스 모델 컴파일 (기본 MSE 사용)
optimizer = Adam(learning_rate=0.001)
source_model.compile(optimizer=optimizer, loss='mean_squared_error')

# WaveBound를 적용한 학습 루프
epochs = 100
batch_size = 32
epsilon = 0.1  # WaveBound의 하이퍼파라미터

for epoch in range(epochs):
    # 소스 모델 학습
    source_model.fit(X_train, y_train, epochs=1, batch_size=batch_size, validation_data=(X_test, y_test), verbose=1)
    
    # 타겟 모델로부터 예측값 생성 (WaveBound 손실 함수에 사용)
    target_pred = target_model.predict(X_train)
    
    # 타겟 모델 업데이트 (EMA 사용)
    update_target_weights(source_model, target_model)
    
    # WaveBound 손실 계산 및 소스 모델 업데이트
    loss = wavebound_loss(y_train, source_model.predict(X_train), target_pred, epsilon=epsilon)
    
    print(f"Epoch {epoch+1}/{epochs}, Loss: {loss}")

# 예측 수행
y_pred = source_model.predict(X_test)

# 성능 평가
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
r_squared = r2_score(y_test, y_pred)

print(f"WaveBound LSTM Model Performance (t-7 시점):")
print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
print(f"MAPE: {mape}%") 
print(f"R-squared: {r_squared}")
```

```python
import os
import pandas as pd
import numpy as np

# 기본 경로 설정
base_dir = r'C:\Users\user\Desktop\Woo LAB\03Fitbit\Fitbit_data'  # 사용자별 데이터가 있는 폴더
output_dir = r'C:\Users\user\Desktop\Woo LAB\03Fitbit\Fitbit_data_processed'  # 처리된 데이터를 저장할 폴더
final_output_file = r'C:\Users\user\Desktop\Woo LAB\03Fitbit\ALL_user_eda_HR.csv'  # 최종 결합 파일 경로

# 저장 폴더가 없다면 생성
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 모든 사용자 데이터를 저장할 리스트
all_data = []

# 사용자별 폴더 탐색
for user_folder in os.listdir(base_dir):
    user_path = os.path.join(base_dir, user_folder)
    if os.path.isdir(user_path):
        file_path = os.path.join(user_path, 'heartrate.csv')
        
        if os.path.exists(file_path):
            # 데이터 로드
            df = pd.read_csv(file_path)
            df['dateTime'] = pd.to_datetime(df['dateTime'])
            df['date'] = df['dateTime'].dt.date  # 날짜만 추출
            
            # 날짜별 그룹화
            grouped = df.groupby('date')
            results = []
            
            for date, group in grouped:
                hr_values = group['value.bpm'].values
                
                # 1. 날짜별 최소, 최대, 표준편차 계산
                hr_min = hr_values.min()
                hr_max = hr_values.max()
                hr_sd = hr_values.std()
                
                # 2. RR 간격 계산
                rr_intervals = 60000 / hr_values
                
                # 3. SDNN 계산
                sdnn = np.std(rr_intervals)
                
                # 4. RMSSD 계산
                rr_diff = np.diff(rr_intervals)
                rmssd = np.sqrt(np.mean(rr_diff ** 2))
                
                # 5. pNN50 계산
                nn50 = np.sum(np.abs(rr_diff) > 50)
                pnn50 = (nn50 / len(rr_diff)) * 100 if len(rr_diff) > 0 else 0
                
                # 결과 저장
                results.append({
                    'date': date,
                    'HR_min': hr_min,
                    'HR_max': hr_max,
                    'HR_sd': hr_sd,
                    'RR': rr_intervals.mean(),  # 평균 RR 간격
                    'SDNN': sdnn,
                    'RMSSD': rmssd,
                    'pNN50': pnn50
                })
            
            # 결과를 데이터프레임으로 변환
            results_df = pd.DataFrame(results)
            results_df['userID'] = user_folder  # 사용자 ID 추가
            
            # 새 경로에 저장
            output_file_path = os.path.join(output_dir, user_folder, 'heartrate_processed.csv')
            os.makedirs(os.path.join(output_dir, user_folder), exist_ok=True)
            results_df.to_csv(output_file_path, index=False)
            print(f"Processed data saved for user {user_folder} at {output_file_path}")
            
            # 모든 사용자 데이터를 리스트에 추가
            all_data.append(results_df)

# 모든 사용자 데이터를 하나의 데이터프레임으로 결합
combined_df = pd.concat(all_data, ignore_index=True)

# 결합된 데이터프레임을 최종 파일로 저장
combined_df.to_csv(final_output_file, index=False)
print(f"All user data combined and saved at {final_output_file}")
```
